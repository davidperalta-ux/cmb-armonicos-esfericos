```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.special import sph_harm_y


theta = np.linspace(0, np.pi, 200)
phi = np.linspace(0, 2*np.pi, 400)

PHI, THETA = np.meshgrid(phi, theta)


Y11 = np.real(sph_harm_y(1, 1, THETA, PHI))
Y21 = np.real(sph_harm_y(2, 1, THETA, PHI))
Y22 = np.real(sph_harm_y(2, 2, THETA, PHI))

combinacion_baja = 1.0*Y11 + 0.7*Y21 + 0.5*Y22


Y31 = np.real(sph_harm_y(3, 1, THETA, PHI))
Y32 = np.real(sph_harm_y(3, 2, THETA, PHI))
Y33 = np.real(sph_harm_y(3, 3, THETA, PHI))

combinacion_alta = 1.0*Y31 + 0.8*Y32 + 0.6*Y33



Y151  = np.real(sph_harm_y(15,  1, THETA, PHI))
Y155  = np.real(sph_harm_y(15,  5, THETA, PHI))
Y1510 = np.real(sph_harm_y(15, 10, THETA, PHI))
Y1515 = np.real(sph_harm_y(15, 15, THETA, PHI))

combinacion_l15 = (1.0*Y151 + 0.8*Y155 + 0.6*Y1510 + 0.4*Y1515)




fig, axs = plt.subplots(1, 3, figsize=(20,6))



im1 = axs[0].imshow(combinacion_baja, extent=[0,360,180,0], aspect='auto', cmap='RdBu_r')
axs[0].set_title(r"Multipolos bajos", fontsize=15)
axs[0].set_xlabel(r"$\phi$ [grados]")
axs[0].set_ylabel(r"$\theta$ [grados]")
plt.colorbar(im1, ax=axs[0], label=r"$\Delta T/T$ relativo")

im2 = axs[1].imshow(combinacion_alta, extent=[0,360,180,0], aspect='auto', cmap='plasma')
axs[1].set_title(r"Multipolos altos", fontsize=15)
axs[1].set_xlabel(r"$\phi$ [grados]")
axs[1].set_ylabel(r"$\theta$ [grados]")
plt.colorbar(im2, ax=axs[1], label=r"$\Delta T/T$ relativo")

im3 = axs[2].imshow(combinacion_l15, extent=[0,360,180,0], aspect='auto', cmap='turbo')
axs[2].set_title(r"Combinación con $\ell = 15$", fontsize=15)
axs[2].set_xlabel(r"$\phi$ [grados]")
axs[2].set_ylabel(r"$\theta$ [grados]")
plt.colorbar(im3, ax=axs[2], label=r"$\Delta T/T$ relativo")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import yn


x = np.linspace(0.0001, 20, 2000)


N0 = yn(0, x)
N1 = yn(1, x)
N2 = yn(2, x)

plt.figure(figsize=(10, 6))

plt.plot(x, N0, label=r"$N_0(x)$")
plt.plot(x, N1, label=r"$N_1(x)$")
plt.plot(x, N2, label=r"$N_2(x)$")

plt.axhline(0, linewidth=0.8)
plt.xlabel(r"$x$")
plt.ylabel(r"$N_n(x)$")
plt.title("Funciones de Neumann $N_0(x)$, $N_1(x)$ y $N_2(x)$")
plt.legend()
plt.grid(True)
plt.ylim(-5, 2)

plt.show()
```
